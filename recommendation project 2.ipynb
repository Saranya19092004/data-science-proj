```python
def similar(df,type,top_n):
    index= type -1
    movie_row= df[index,:]
    magnitude= np.sqrt(np.einsum('ij, ij -> i',df,df))
    similarity= np.dot(movie_row,df.T) / (magnitude[index]*magnitude)
    sort_indexes= np.argsort(-similarity)
    return sort_indexes[:top_n]
```
