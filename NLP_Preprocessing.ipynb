TOKENIZED


```python
import pandas as pd
from nltk.tokenize import word_tokenize
import nltk

# Download the NLTK tokenizer models (only need to run once)
nltk.download('punkt_tab') # Changed from 'punkt' to 'punkt_tab' as per error message

# Load CSV file using pandas, specifying tab as separator and no header
df = pd.read_csv('/content/Restaurant_Reviews.tsv', sep='\t', header=None, names=['text', 'sentiment'])

# Apply tokenization to each row in 'text' column
df['tokens'] = df['text'].apply(word_tokenize)

# Save the DataFrame with tokens back to CSV if needed
df.to_csv('tokenized_output.csv', index=False)

# Print first few rows with tokens
print(df[['text', 'tokens']].head())
```

STOP WORDS


```python
nlp=spacy.load("en_core_web_sm")

# Define a function to process a single text entry and find stop words
def get_spacy_stop_words(text_entry):
    if isinstance(text_entry, str):
        doc = nlp(text_entry)
        # Collect only the text of tokens that are stop words
        return [token.text for token in doc if token.is_stop]
    return [] # Return empty list for non-string inputs

# Apply this function to the 'text' column and store the results in a new column
df['spacy_stop_words'] = df['text'].apply(get_spacy_stop_words)

# Print the original text and the identified stop words for the first few entries
print(df[['text', 'spacy_stop_words']].head())
```

STEMMING

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
```

```python
print("Stemming examples (first 5 reviews):")
# Iterate over each list of stop words in the 'spacy_stop_words' column
for i, stop_word_list in enumerate(df['spacy_stop_words']):
    if i >= 5: # Limit output to first 5 reviews for demonstration
        break
    print(f"Review {i}:")
    if isinstance(stop_word_list, list):
        for word_to_stem in stop_word_list:
            print(f"  {word_to_stem} | {stemmer.stem(word_to_stem)}")
    else:
        # Handle cases where an entry might not be a list (e.g., if it's empty or malformed)
        print(f"  Non-list entry: {stop_word_list}")
    print("-" * 20)
```

LEMMATIZATION

```python
print("Lemmatization examples for Spacy stop words (first 5 reviews):")
# Iterate over each list of stop words in the 'spacy_stop_words' column
for i, stop_word_list in enumerate(df['spacy_stop_words']):
    if i >= 5: # Limit output to first 5 reviews for demonstration
        break
    print(f"Review {i}:")
    if isinstance(stop_word_list, list):
        for word_to_lemmatize in stop_word_list:
            # Process each individual word with nlp to get its lemma
            doc_word = nlp(word_to_lemmatize)
            print(f"  {word_to_lemmatize} | {doc_word[0].lemma_}")
    else:
        print(f"  Non-list entry: {stop_word_list}")
    print("-" * 20)
```

VECTORIZER -----  BAG OF WORDS

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import spacy

# Ensure df is correctly loaded and 'spacy_stop_words' is present
# Reload df in case it was modified or reset
df = pd.read_csv('/content/tokenized_output.csv')

# Re-load spacy model if not already loaded, and re-create 'spacy_stop_words'
nlp = spacy.load("en_core_web_sm")

def get_spacy_stop_words(text_entry):
    if isinstance(text_entry, str):
        doc = nlp(text_entry)
        return [token.text for token in doc if token.is_stop]
    return []

df['spacy_stop_words'] = df['text'].apply(get_spacy_stop_words)

# Convert the list of stop words in each row into a single string for CountVectorizer
# This ensures each entry is a string, which CountVectorizer expects as a 'document'
df_for_bow = df.copy() # Create a copy to avoid modifying the original df's column type directly if not intended
df_for_bow['spacy_stop_words_str'] = df_for_bow['spacy_stop_words'].apply(lambda x: ' '.join(x) if isinstance(x, list) else '')

# Initialize CountVectorizer (Bag of Words model)
vectorizer = CountVectorizer()

# Fit and transform the text data (the joined stop words) to BoW matrix
bow_matrix = vectorizer.fit_transform(df_for_bow['spacy_stop_words_str'])

# Convert BoW matrix to DataFrame for better visualization
bow_df = pd.DataFrame(bow_matrix.toarray(), columns=vectorizer.get_feature_names_out())

print("Bag of Words for Spacy Stop Words (first 5 rows):")
print(bow_df.head())
```
