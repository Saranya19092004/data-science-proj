```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D,MaxPooling2D,Flatten
import numpy as np
```

```python
import zipfile
zip_ref=zipfile.ZipFile("/content/car vs bike.zip",'r')
zip_ref.extractall()
zip_ref.close()
```

```python
import tensorflow.keras as keras
train=keras.utils.image_dataset_from_directory(
    directory='/content/Car-Bike-Dataset',
    labels='inferred',
    label_mode='int',
    batch_size=32,
    image_size=(256,256)
)
```

```python
test=keras.utils.image_dataset_from_directory(
    directory='/content/Car-Bike-Dataset',
    labels='inferred',
    label_mode='int',
    batch_size=32,
    image_size=(256,256)
)
```

```python
for images, labels in train.take(1):
  print(images.shape)
  print(labels.shape)
```

```python
def process(image,labels):
  image=image/255
  return image,labels

  train=train.map(process)
  test=test.mao(process)
```

```python
model=Sequential([
    Conv2D(32,(3,3),activation='relu',input_shape=(256,256,3)),
    MaxPooling2D((2,2)),
    Conv2D(64,(3,3),activation='relu'),
    MaxPooling2D((2,2)),
    Conv2D(128,(3,3),activation='relu'),
    MaxPooling2D((2,2)),
    Conv2D(128,(3,3),activation='relu'),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(512,activation='relu'),
    Dense(1,activation='sigmoid')

])
```

```python
model=Sequential()
model.add(Conv2D(32,(3,3),activation='relu',input_shape=(256,256,3))),
model.add(MaxPooling2D((2,2))),
model.add(Conv2D(64,(3,3),activation='relu')),
model.add(MaxPooling2D((2,2))),
model.add(Conv2D(128,(3,3),activation='relu')),
model.add(MaxPooling2D((2,2))),
model.add(Conv2D(128,(3,3),activation='relu')),
model.add(MaxPooling2D((2,2))),
model.add(Flatten()),
model.add(Dense(512,activation='relu')),
model.add(Dense(1,activation='sigmoid'))
```

```python
train_images_ds = train.map(lambda image, label: image)
val_images_ds = test.map(lambda image, label: image)

feature_train = model.predict(train_images_ds)
feature_test = model.predict(val_images_ds)
```

```python
import tensorflow as tf

# Apply error handling to the base datasets first
train = train.apply(tf.data.experimental.ignore_errors())
test = test.apply(tf.data.experimental.ignore_errors())

# Extract images from the (image, label) datasets
train_images_ds = train.map(lambda image, label: image)
val_images_ds = test.map(lambda image, label: image)

# Apply normalization (divide by 255.0)
train_images_ds = train_images_ds.map(lambda image: image / 255.0)
val_images_ds = val_images_ds.map(lambda image: image / 255.0)

# Add caching and prefetching for improved robustness and performance
train_images_ds = train_images_ds.cache().prefetch(buffer_size=tf.data.AUTOTUNE)
val_images_ds = val_images_ds.cache().prefetch(buffer_size=tf.data.AUTOTUNE)

feature_train = model.predict(train_images_ds)
feature_test = model.predict(val_images_ds)
```

```python
from sklearn.tree import DecisionTreeClassifier

train_labels = np.concatenate([label.numpy() for image, label in train], axis=0)
test_labels = np.concatenate([label.numpy() for image, label in test], axis=0)

dc=DecisionTreeClassifier()
dc.fit(feature_train, train_labels)
y_pred = dc.predict(feature_test)
```

```python
dc=DecisionTreeClassifier()
dc.fit(feature_test, test_labels)
y_pred = dc.predict(feature_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(test_labels,y_pred))
```

```python
print(accuracy_score(train_labels,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
ra=RandomForestClassifier()
ra.fit(feature_train, train_labels)
y_pred=ra.predict(feature_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(test_labels,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
ra=RandomForestClassifier(n_estimators=500,criterion='gini',max_depth=5,min_samples_split=8,random_state=23,min_samples_leaf=4,max_features='sqrt')
ra.fit(feature_train, train_labels)
y_pred=ra.predict(feature_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(test_labels,y_pred))
```
